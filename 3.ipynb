```python
import requests

url = "http://universities.hipolabs.com/search"

Response = requests.get(url)
Response
```

```python
data = Response.json()
data
```

```python
for d in data:
  print(d)
```

```python
import csv
with open("uni_data.csv",'w',newline='') as file :


  writer = csv.DictWriter(file ,fieldnames=['Name','Country','Web page'])
  writer.writeheader()

  for d in data:
    Country = d['country']
    Name = d['name']
    Web_page = d['web_pages']

    writer.writerow({ "Name" : Name ,
                     "Country" : Country ,
                      'Web page' : Web_page
                      })
```
